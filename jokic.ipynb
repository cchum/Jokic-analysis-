```python
import pandas as pd
```

```python
data = pd.read_csv('jokic.csv')
```

```python
data.shape
```

```python
data.head(12)
```

```python
data.drop(index=0, inplace=True)
data.drop(index=11, inplace=True)
```

```python
data.reset_index(drop=True)
```

```python
data.info()
```

```python
data = data.drop(columns=['Team', 'Lg', 'Pos'])
```

```python
data['Awards'] = data['Awards'].fillna('No awards')
```

```python
data.info()
```

```python
data
```

```python
for col in data.columns:
    if col not in ['Awards', 'Season']:
        data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
data.dtypes
```

```python
data['Has award'] = data['Awards'].apply(lambda x: 0 if x == 'No awards' else 1)
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['Season'], data['PTS'], marker='o', label='Points Per Game')
plt.plot(data['Season'], data['AST'], marker='o', label='Assist Per Game')
plt.plot(data['Season'], data['TRB'], marker='o', label='Rebounds Per Game')

plt.xticks(rotation=45)
plt.title('Nikola Jokic Scoring, Passing, Rebounding Progression')
plt.legend()
plt.show()
plt.clf()
```

```python
plt.scatterplot(data['MP'], data['PTS'],c=data['Has award'], cmap='coolwarm', s=100)
plt.xlabel('Minutes Per Game')
plt.ylabel('Points Per Game')
plt.title('Effe')
```
